```python
import numpy as np
import pandas as pd
import matplotlib as mpl 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Import Data 

data = pd.read_csv("globalterrorismdb_0718dist.csv",encoding ="latin1")
df =pd.DataFrame(data)
print("Data imported")
df.head(10)
```

```python
df.info()
```

```python
df.columns
```

```python
# to see all name

for i in df.columns:
    print(i,end =",")
```

```python
#clean data -as we need only some clomuns to study data
df =df[["iyear","imonth","iday","country_txt","region_txt",
       "provstate","city","latitude","longitude","summary",
       "attacktype1_txt","targtype1_txt","gname","motive",
       "]]
df.head()
```

```python
df.rename(columns ={"iyear":"Year","imonth":"Month","iday":"Day","country_txt":"Country","region_txt":"Region",
       "provstate":"Province/State","city":"City","latitude":"Latitude","longitude":"Longitude","summary":"Summary",
       "attacktype1_txt":"Attack Type","targtype1_txt":"Target Type","gname":"Group Name","motive" :"Motive",
       "weaptype1_txt":"weapon Type","nkill":"Killed","nwound":"Wounded","addnotes":"Add notes"},inplace =True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["Killed"] = df["Killed"].fillna(0)
df["Wounded"] = df["Wounded"].fillna(0)
df["Casualty"] = df["Killed"] +df["Wounded"]
```

```python
df.describe()
```

```python
#visualizing data

attacks =df["Year"].value_counts(dropna =False).sort_index().to_frame().reset_index().rename(columns ={"index":"Year","Year":"Attack"}).set_index("Year")
```

```python
attacks.head()
```

```python
attacks.plot(kind ="bar",color ="cornflowerblue",figsize=(15,6),fontsize =13)
plt.title("Timeline of Attack",fontsize =15)
plt.xlabel("Year",fontsize =15)
plt.ylabel("MNo of Attack",fontsize =15)
plt.show()
```

```python
yc =df[["Year","Casualty"]].groupby("Year").sum()
yc.head()
```

```python
yc.plot(kind ="bar",color ="cornflowerblue",figsize=(15,6))
plt.title("Year Wise Casualties",fontsize =15)
plt.xlabel("Year",fontsize =15)
plt.xticks(fontsize =12)
plt.ylabel("NO of Casulaties",fontsize =15)
plt.show()
```

```python
yc =df[["Year","Killed"]].groupby("Year").sum()
yc.head()
```

```python
yk =df[["Year","Wounded"]].groupby("Year").sum()
yk.head()
```

```python
yc.plot(kind ="bar",color ="cornflowerblue",figsize=(15,6))
plt.title("Year Wise Killed",fontsize =15)
plt.xlabel("Year",fontsize =15)
plt.xticks(fontsize =12)
plt.ylabel("NO of Killed",fontsize =15)
plt.show()
```

```python
yk.plot(kind ="bar",color ="cornflowerblue",figsize=(15,6))
plt.title("Year Wise Wounded",fontsize =15)
plt.xlabel("Year",fontsize =15)
plt.xticks(fontsize =12)
plt.ylabel("NO of wounded",fontsize =15)
plt.show()
```

```python
# see by region

reg =pd.crosstab(df.Year,df.Region)
reg.head()
```

```python
reg.plot(kind ="area",stacked =False,alpha =0.5,figsize =(20,10))
plt.title("Region Wise Attacks ",fontsize =20)
plt.xlabel("Year",fontsize =15)
plt.xticks(fontsize =12)
plt.ylabel("NO of Attacks",fontsize =15)
plt.show()
```

```python
regt = reg.transpose()
regt["Total"] = regt.sum(axis =1)
ra = regt["Total"].sort_values(ascending =False)
ra
```

```python
ra.plot(kind ="bar",figsize =(15,6))
plt.title("Total no of attacks in each Region from 1970-2017")
plt.xlabel("Region")
plt.ylabel("No of Attacks")
plt.show()
```

```python


# Total Casualties Kiled +wounded in each ergin
rc = df[["Region","Casualty"]].groupby("Region").sum().sort_values(by="Casualty",ascending =False)
rc
```

```python
rc.plot(kind ="bar",color ="cornflowerblue",figsize=(15,6))
plt.title("Region Wise Wounded",fontsize =15)
plt.xlabel("Year",fontsize =15)
plt.xticks(fontsize =12)
plt.ylabel("NO of Casualties",fontsize =15)
plt.show()
```
